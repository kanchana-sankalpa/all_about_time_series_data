```python
# -*- coding: utf-8 -*-
"""
Join two datasets within a folder

Created on Wed Jan  6 12:13:02 2021

@author: kanchana
"""
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import glob
import matplotlib.dates as mdates


cwd = os.getcwd()  # Get the current working directory (cwd)
files = os.listdir(cwd)  # Get all the files in that directory
print("Files in %r: %s" % (cwd, files))

path = cwd+'/QLD_open_data/sampled_30' # use your path
all_files = glob.glob(path + "/*.csv")
df = pd.DataFrame() 

for idx, filename in enumerate(all_files):
    if idx == 0:
        print('index 0')
        
        df1 = pd.read_csv(filename, index_col=False, header=0)
        df1.dropna(axis = 0, how ='all', inplace = True)
        
        filename_w_ext = os.path.basename(filename)
        exact_filename, file_extension = os.path.splitext(filename_w_ext)
        print('exact_filename')
        print(exact_filename)
        
        
        print(df1.shape)
        print(df1.dtypes)

        
        df1['TIMESTAMP'] = pd.to_datetime(df1['TIMESTAMP'],  format="%Y-%m-%d %H:%M:%S" , errors='coerce')
        df1 = df1.set_index('TIMESTAMP')
        
        print('df1')
        print(df1.head(10))
        df1.sort_index()
        df=df1
        
    else:
        df2 = pd.read_csv(filename, index_col=False, header=0)
        df2.dropna(axis = 0, how ='all', inplace = True)
        
        filename_w_ext = os.path.basename(filename)
        exact_filename, file_extension = os.path.splitext(filename_w_ext)
        print('exact_filename')
        print(exact_filename)
        
        
        print(df2.shape)
        print(df2.dtypes)

        
        df2['TIMESTAMP'] = pd.to_datetime(df2['TIMESTAMP'],  format="%Y-%m-%d %H:%M:%S" , errors='coerce')
        df2 = df2.set_index('TIMESTAMP')
        
        print('df2')
        print(df2.head(10))
        df2.sort_index()
        
        results = pd.DataFrame()
        
        if(df.shape[0] < df2.shape[0]):
            print('df.shape[0] < df2.shape[0]')
            results = pd.merge(df2.sort_index(), df.sort_index(), on='TIMESTAMP', how='outer')
        else:
            print('--other_condition ---')
            results = pd.merge(df.sort_index(), df2.sort_index(), on='TIMESTAMP', how='outer')
           
        results=results.sort_index()
        print('----results-----')
        print(results.shape)
        df=results
        
print('saving file')
df.to_csv(cwd+'/QLD_open_data/combined_buff_burnett_sampled_30.csv') 
```
